```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list_alpha = [0.05, 0.01, 0.005]
values = [[], [], []]
max_num = 13
for i in range(1, max_num + 1):
    for j, alpha in enumerate(list_alpha):
        values[j].append(1 - (1 - alpha) ** i)

import itertools
marker = itertools.cycle(('+', '.', '*', 'v'))

for i, alpha in enumerate(list_alpha):
    plt.plot(np.arange(1, max_num + 1), values[i], label=f'Î±={alpha}', marker=next(marker))

plt.legend()
plt.ylim((-0.1, 0.6))
y_labels = [0.0, 0.05]
y_labels.extend(np.arange(0.1, 0.6, 0.1).tolist())
plt.yticks(y_labels)
plt.gca().get_yticklabels()[1].set_color("red")
plt.xticks(np.arange(1, max_num + 1))
plt.axhline(y=0.05, color='r', linestyle='--')
plt.ylabel('Experiment-wise Error')
plt.xlabel('Number of Comparisons')
plt.savefig('experiment-wise_error_curve.pdf')
```
